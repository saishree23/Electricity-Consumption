## **Getting started**
</div>

The document you are reading is not a static web page, but an interactive environment called a **Colab notebook** that lets you write and execute code.

For example, here is a **code cell** with a short Python script that computes a value, stores it in a variable, and prints the result:

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.preprocessing import MinMaxScaler
```

```

# Load the dataset with specified data types and handle non-numeric values
data = pd.read_csv('/household_power_consumption.txt', delimiter=';', dtype=str, parse_dates={'DateTime': ['Date', 'Time']})

# Replace non-numeric values with NaN
data.replace('?', np.nan, inplace=True)

# Convert numeric columns to float type
numeric_cols = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
data[numeric_cols] = data[numeric_cols].astype(float)

# Handle missing values
data.dropna(inplace=True)

# Plot Global Active Power over time
plt.figure(figsize=(12, 6))
plt.plot(data['DateTime'], data['Global_active_power'], color='blue')
plt.title('Electricity Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Global Active Power')
plt.grid(True)
plt.show()


# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Step 3: Time Series Forecasting Models

# Split the dataset into train and test sets
train_size = int(len(data) * 0.8)
train, test = data.iloc[:train_size], data.iloc[train_size:]

# ARIMA model
def arima_model(train, test):
    model = ARIMA(train['Global_active_power'], order=(5,1,0))
    model_fit = model.fit()
    predictions = model_fit.forecast(steps=len(test))
    return predictions

arima_predictions = arima_model(train, test)

# LSTM model
def lstm_model(train, test):
    scaler = MinMaxScaler()
    scaler.fit(train[['Global_active_power']])
    scaled_train = scaler.transform(train[['Global_active_power']])
    scaled_test = scaler.transform(test[['Global_active_power']])

    X_train, y_train = scaled_train[:-1], scaled_train[1:]
    X_test, y_test = scaled_test[:-1], scaled_test[1:]

    X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

    model = Sequential()
    model.add(LSTM(50, activation='relu', input_shape=(1, 1)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=100, verbose=0)

    lstm_predictions = model.predict(X_test)
    lstm_predictions = scaler.inverse_transform(lstm_predictions)
    return lstm_predictions

lstm_predictions = lstm_model(train, test)

# Step 4: Feature Engineering
# (Not included in this code snippet)

# Step 5: Model Evaluation and Tuning

# Evaluate ARIMA model
arima_rmse = np.sqrt(mean_squared_error(test['Global_active_power'], arima_predictions))
print('ARIMA RMSE:', arima_rmse)

# Evaluate LSTM model
lstm_rmse = np.sqrt(mean_squared_error(test['Global_active_power'].iloc[1:], lstm_predictions))
print('LSTM RMSE:', lstm_rmse)
```
